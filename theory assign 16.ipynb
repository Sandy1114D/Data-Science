1. Here's an example of how to create `years_list`:

```python
years_list = [1990, 1991, 1992, 1993, 1994, 1995]
```

2. You were born in 1990, so your third birthday would have been in 1993, which is the third year in `years_list`.

4. Here's an example of how to create the `things` list:

```python
things = ["mozzarella", "cinderella", "salmonella"]
```

5. To capitalize the element in `things` that refers to a person, we can use the `index` method to find its position in the list, and then assign the capitalized version of the string to that position:

```python
things[1] = things[1].capitalize()
print(things)
```

This would output: `["mozzarella", "Cinderella", "salmonella"]`. Yes, it changed the element in the list.

6. Here's an example of how to create the `surprise` list:

```python
surprise = ["Groucho", "Chico", "Harpo"]
```

7. To lowercase the last element of `surprise`, reverse it, and then capitalize it, we can use string methods and list indexing:

```python
surprise[-1] = surprise[-1].lower()[::-1].capitalize()
print(surprise)
```

This would output: `["Groucho", "Chico", "Oprah"]`.

8. Here's an example of how to create the `e2f` dictionary:

```python
e2f = {"dog": "chien", "cat": "chat", "walrus": "morse"}
print(e2f)
```

This would output: `{"dog": "chien", "cat": "chat", "walrus": "morse"}`.

9. The French word for "walrus" in `e2f` is "morse".

10. Here's an example of how to create the `f2e` dictionary from `e2f` using the `items` method:

```python
f2e = {french: english for english, french in e2f.items()}
print(f2e)
```

This would output: `{"chien": "dog", "chat": "cat", "morse": "walrus"}`.

11. To print the English version of the French word "chien" using `f2e`, we can use dictionary indexing:

```python
print(f2e["chien"])
```

This would output: `"dog"`.

12. To make a set of English words from the keys in `e2f`, we can use the `set` function:

```python
english_words = set(e2f.keys())
print(english_words)
```

This would output: `{"dog", "cat", "walrus"}`.
