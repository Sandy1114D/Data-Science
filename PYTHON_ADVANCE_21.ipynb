ANS OF ABOVE QUESTIONS ARE AS FOLLOW

Q1. A probability distribution is a function that describes the likelihood of different outcomes in a random event. While the specific outcome of any individual event may be unpredictable, the distribution can provide information about the likelihood of various outcomes occurring over a large number of trials.

Q2. True random numbers are generated by physical processes that are inherently unpredictable, such as radioactive decay. Pseudo-random numbers are generated by algorithms that use mathematical formulas to produce numbers that have statistical properties similar to true random numbers. While they are not truly random, pseudo-random numbers are considered "good enough" for many purposes.

Q3. The two main factors that influence the behaviour of a normal probability distribution are the mean and the standard deviation. The mean determines the center of the distribution, while the standard deviation determines the spread or variability of the distribution.

Q4. A real-life example of a normal distribution is the distribution of heights in a population. The majority of people are of average height, with progressively fewer people as you move towards the extremes of very tall or very short.

Q5. In the short term, a probability distribution may not follow its expected pattern, as random events can produce unexpected outcomes. However, as the number of trials increases, the distribution will tend to more closely resemble its expected pattern, as the effect of random variation becomes less significant.

Q6. The random.shuffle function can shuffle lists.

Q7. The math package in Python provides functions in several general categories, including:

- Trigonometric functions: sin, cos, tan, asin, acos, atan
- Exponential and logarithmic functions: exp, log, log10
- Hyperbolic functions: sinh, cosh, tanh, asinh, acosh, atanh
- Constants: pi, e

Q8. Exponentiation and logarithms are inverse operations. If x^y = z, then log_x(z) = y. In other words, taking the logarithm of a number with a given base will tell you the exponent to which you must raise the base to obtain that number.

Q9. Python supports three logarithmic functions: 

- math.log(x) returns the natural logarithm (base e) of x.
- math.log10(x) returns the base-10 logarithm of x.
- math.log2(x) returns the base-2 logarithm of x.
